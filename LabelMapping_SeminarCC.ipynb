```python
from datasets import load_dataset, Features, ClassLabel, Value

# Load the dataset
dataset = load_dataset("climatebert/climate_sentiment")

# Define the new feature schema
features = dataset['train'].features
features['mapped_label'] = Value('string')

# Function to add a new column with the mapped string labels
def add_label_column(example):
    label_map = {0: "REFUTES", 1: "NOT_ENOUGH_INFO", 2: "SUPPORTS"}
    example['mapped_label'] = label_map[example['label']]
    return example

# Apply the function to add the new label column with the updated features
mapped_dataset = dataset.map(add_label_column, features=features)

# Example to check the result
# print(mapped_dataset['train'][0])

# Save the dataset to a text file
output_file = "climate_sentiment_mapped.txt"

with open(output_file, 'w') as f:
    for example in mapped_dataset['train']:
        f.write(f"{example['text']}\t{example['label']}\t{example['mapped_label']}\n")

print(f"Dataset saved to {output_file}")
```
